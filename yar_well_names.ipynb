```python
import pandas as pd
```

```python
def func(x:str)->list:
    a = x.split("_")
    return [a[0], f"{a[1]}_{a[2]}"]
```

```python
well_names = pd.read_excel("yar_names.xlsx")
well_project = pd.read_excel("yar_project.xlsx")
well_names
```

```python
lst = list(well_project["Name"].apply(func))
well_project = pd.DataFrame(lst, columns=["Field", "Well"])
```

```python
well_project
```

```python
itog = pd.merge(left=well_names, right=well_project, how="left", left_on="Well", right_on="Well")
itog = itog.astype({"Well":"string", "Field":"string" })
itog["Well"] = itog["Field"] + "_" + itog["Well"]
itog = itog.drop(columns="Field")
itog.to_excel("yar_wells_itog.xlsx")
```
