```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("fault.xlsx")
fault=[]
stics = pd.DataFrame({
    'X':[],
    'Y':[],
    'Z':[]
})
z_top = 1750
z_bot = 1950
data
```

```python
pol_end = data[data["Z"]==999]
pol_end["pol"] = range(0,len(pol_end))
data["pol"] = 0
pol_end["index"] = list(pol_end.index)
pol_end
```

```python
for i in range(0, len(pol_end)):
    if i==0:
        data.loc[i:pol_end["index"].iloc[i], "pol"] = 0
    else:
        data.loc[int(pol_end.iloc[i-1]["index"] + 1) : int(pol_end.iloc[i]["index"]), "pol"] = i # i = 1..14

data = data[data["X"]!=999]
```

```python
stics.loc[len(stics)] = [1,2,3]
stics = pd.DataFrame({
    'X':[],
    'Y':[],
    'Z':[],
    'pol':[]
})
stics
```

```python
up = len(pol_end)
for i in range(0, up):
    for el in data[data["pol"]==i].values:
        stics.loc[len(stics)] = [el[0], el[1], z_top, el[3]]
        stics.loc[len(stics)] = [el[0], el[1], z_bot, el[3]]
        stics.loc[len(stics)] = [999, 999, 999, el[3]]
```

```python
for i in range(len(pol_end)):
    stics[stics["pol"]==i].to_csv(f"stick{i}.txt", sep=" ", index=False, columns=["X","Y","Z"])
```
